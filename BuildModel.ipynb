```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')


```

```python
data_ARIMA = pd.read_csv('results.csv', sep=',', header=0, parse_dates=True, index_col=0 )
data_ARIMA
```

```python
data_ARIMA_float = data_ARIMA[:].astype(np.float)
data_ARIMA_float.index = pd.to_datetime(data_ARIMA_float.index)
data_ARIMA_float
```

```python
data_ARIMA_float.index
```

```python
train = data_ARIMA_float.loc[:'2017-12-01',]
test = data_ARIMA_float.loc['2018-01-01':,]
```

```python
train.index
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(train, model='multiplicative')
fig = result.plot()
plt.show()
```

```python
from pmdarima.arima import auto_arima
Rstepwise_model = auto_arima(train.loc[:,'Rcnt'], start_p=1, start_q=1,
                           max_p=3, max_q=3, m=12,
                           start_P=0, seasonal=True,
                           d=1, D=1, trace=True,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)
print(Rstepwise_model.aic())
```

```python
Rstepwise_model.fit(train.loc[:,'Rcnt'])
Rforecast = Rstepwise_model.predict(n_periods=12)
Rforecastdf = pd.DataFrame(Rforecast,index = test.index,columns=['RPredict'])
```

```python
pd.concat([test.loc[:,'Rcnt'],Rforecastdf],axis=1).plot()
```

```python
Pstepwise_model = auto_arima(train.loc[:,'Pcnt'], start_p=1, start_q=1,
                           max_p=3, max_q=3, m=12,
                           start_P=0, seasonal=True,
                           d=1, D=1, trace=True,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)
print(Pstepwise_model.aic())
```

```python
Pstepwise_model.fit(train.loc[:,'Pcnt'])
Pforecast = Pstepwise_model.predict(n_periods=12)
Pforecastdf = pd.DataFrame(Pforecast,index = test.index,columns=['PPredict'])
```

```python
pd.concat([test.loc[:,'Pcnt'],Pforecastdf],axis=1).plot()
```

```python
from sklearn.externals import joblib  # for persistence
joblib.dump(Pstepwise_model, 'pmodel.pkl', compress=3)
joblib.dump(Rstepwise_model, 'rmodel.pkl', compress=3)
```

```python
parima_loaded = joblib.load('pmodel.pkl')
rarima_loaded = joblib.load('rmodel.pkl')
```

```python
Pforecast6year = parima_loaded.predict(n_periods=72)
Pforecast6year
```

```python
Rforecast6year = rarima_loaded.predict(n_periods=72)
Rforecast6year
```

```python
import datetime
from dateutil.relativedelta import relativedelta
start = datetime.datetime.strptime("2018-01-01", "%Y-%m-%d")
date_list = [start + relativedelta(months=+x) for x in range(0, 72)]
date_list
```

```python
Rfuture_forecast = pd.DataFrame(Rforecast6year, index=date_list,columns=['RPredict'])
Pfuture_forecast = pd.DataFrame(Pforecast6year, index=date_list,columns=['PPredict'])
```

```python
#remove 2018
Rfuture_forecast = Rfuture_forecast.iloc[12:]
```

```python
#remove 2018
Pfuture_forecast = Pfuture_forecast.iloc[12:]
Pfuture_forecast
```

```python
data_final = pd.concat([data_ARIMA,Rfuture_forecast, Pfuture_forecast],axis=1)
data_final
```

```python
fig = data_final.plot(figsize=(15, 15)).get_figure()
```
